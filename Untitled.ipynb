```python
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.figsize"] = (12, 6)

BASE_DIR = Path.cwd()
DATA_DIR = BASE_DIR / "data" / "azure"

AZURE_FILES = {
    "telemetry": DATA_DIR / "PdM_telemetry.csv",
    "errors": DATA_DIR / "PdM_errors.csv",
    "maint": DATA_DIR / "PdM_maint.csv",
    "failures": DATA_DIR / "PdM_failures.csv",
    "machines": DATA_DIR / "PdM_machines.csv",
}

print(f"Repertoire de travail : {BASE_DIR}")
print(f"Repertoire des donnees : {DATA_DIR}")
```

```python
missing_files = [str(path) for path in AZURE_FILES.values() if not path.exists()]

if missing_files:
    raise FileNotFoundError(
        "Fichiers Azure manquants. Ajoute les CSV suivants dans data/azure/:\n- "
        + "\n- ".join(missing_files)
    )

print("Tous les fichiers Azure requis sont presents.")
```

## 1. Chargement des donnees

On charge les tables principales du projet Azure et on convertit les colonnes temporelles au bon format.


```python
telemetry = pd.read_csv(AZURE_FILES["telemetry"], parse_dates=["datetime"])
errors = pd.read_csv(AZURE_FILES["errors"], parse_dates=["datetime"])
maint = pd.read_csv(AZURE_FILES["maint"], parse_dates=["datetime"])
failures = pd.read_csv(AZURE_FILES["failures"], parse_dates=["datetime"])
machines = pd.read_csv(AZURE_FILES["machines"])

datasets = {
    "telemetry": telemetry,
    "errors": errors,
    "maint": maint,
    "failures": failures,
    "machines": machines,
}

for name, df in datasets.items():
    print(f"{name:<10} -> {df.shape[0]:>8,} lignes x {df.shape[1]} colonnes")
```

```python
def dataset_overview(df, name, n=5):
    print(f"\n===== {name.upper()} =====")
    display(df.head(n))
    info_df = pd.DataFrame({
        "dtype": df.dtypes.astype(str),
        "missing_count": df.isna().sum(),
        "missing_pct": (df.isna().mean() * 100).round(2),
        "n_unique": df.nunique(),
    })
    display(info_df)

for name, df in datasets.items():
    dataset_overview(df, name)
```

## 2. Resume statistique des variables numeriques

Cette partie permet de reperer les ordres de grandeur, la dispersion et de premiers signaux de valeurs extremes.


```python
telemetry.describe().T
```

```python
missing_summary = (
    telemetry.isna()
    .mean()
    .mul(100)
    .round(2)
    .rename("missing_pct")
    .reset_index()
    .rename(columns={"index": "column"})
)

display(missing_summary)

plt.figure(figsize=(10, 4))
sns.barplot(data=missing_summary, x="column", y="missing_pct", color="steelblue")
plt.title("Taux de valeurs manquantes par variable - Telemetry")
plt.xlabel("Variable")
plt.ylabel("Pourcentage manquant")
plt.xticks(rotation=30)
plt.show()
```

## 3. Courbes de capteurs par machine

On observe l'evolution temporelle des capteurs sur quelques machines pour detecter des comportements anormaux ou des trajectoires de degradation.


```python
sensor_cols = ["volt", "rotate", "pressure", "vibration"]
sample_machines = sorted(telemetry["machineID"].dropna().unique())[:5]

fig, axes = plt.subplots(len(sensor_cols), 1, figsize=(15, 16), sharex=True)

for idx, sensor in enumerate(sensor_cols):
    ax = axes[idx]
    for machine_id in sample_machines:
        tmp = telemetry.loc[telemetry["machineID"] == machine_id].sort_values("datetime")
        ax.plot(tmp["datetime"], tmp[sensor], label=f"Machine {machine_id}", alpha=0.8)
    ax.set_title(f"Evolution temporelle de {sensor} pour 5 machines")
    ax.set_ylabel(sensor)
    ax.legend(ncol=5, fontsize=9, loc="upper right")

axes[-1].set_xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
failure_events = failures.copy()
failure_events["failure_event"] = 1

telemetry_with_failures = telemetry.merge(
    failure_events[["datetime", "machineID", "failure_event"]],
    on=["datetime", "machineID"],
    how="left",
)
telemetry_with_failures["failure_event"] = telemetry_with_failures["failure_event"].fillna(0).astype(int)

machine_focus = sample_machines[0]
focus_df = telemetry_with_failures.loc[
    telemetry_with_failures["machineID"] == machine_focus
].sort_values("datetime")

fig, axes = plt.subplots(len(sensor_cols), 1, figsize=(15, 16), sharex=True)

for idx, sensor in enumerate(sensor_cols):
    ax = axes[idx]
    ax.plot(focus_df["datetime"], focus_df[sensor], color="tab:blue")
    failure_dates = focus_df.loc[focus_df["failure_event"] == 1, "datetime"]
    for failure_date in failure_dates:
        ax.axvline(failure_date, color="crimson", linestyle="--", alpha=0.7)
    ax.set_title(f"{sensor} - Machine {machine_focus} avec dates de panne")
    ax.set_ylabel(sensor)

axes[-1].set_xlabel("Date")
plt.tight_layout()
plt.show()
```

## 4. Valeurs aberrantes

On combine une lecture visuelle par boxplots et un comptage simple par la methode IQR pour identifier les variables les plus sensibles aux valeurs extremes.


```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for ax, col in zip(axes, sensor_cols):
    sns.boxplot(data=telemetry, y=col, ax=ax, color="lightblue")
    ax.set_title(f"Boxplot - {col}")

plt.tight_layout()
plt.show()

outlier_rows = []
for col in sensor_cols:
    q1 = telemetry[col].quantile(0.25)
    q3 = telemetry[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    mask = (telemetry[col] < lower) | (telemetry[col] > upper)
    outlier_rows.append({
        "variable": col,
        "lower_bound": round(lower, 3),
        "upper_bound": round(upper, 3),
        "outlier_count": int(mask.sum()),
        "outlier_pct": round(mask.mean() * 100, 3),
    })

outlier_summary = pd.DataFrame(outlier_rows).sort_values("outlier_pct", ascending=False)
display(outlier_summary)
```

## 5. Heatmap de correlations

On enrichit la table de telemetry avec les caracteristiques machine et un indicateur de panne pour obtenir une vue plus utile a l'interpretation.


```python
azure_eda = telemetry.merge(machines, on="machineID", how="left")
azure_eda = azure_eda.merge(
    failure_events[["datetime", "machineID", "failure_event"]],
    on=["datetime", "machineID"],
    how="left",
)
azure_eda["failure_event"] = azure_eda["failure_event"].fillna(0).astype(int)

corr = azure_eda.select_dtypes(include=np.number).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Heatmap de correlation - Azure Predictive Maintenance")
plt.show()
```

## 6. Synthese automatique

Cette cellule produit une synthese courte directement reutilisable dans le rapport ou pour la restitution orale de la semaine 1.


```python
top_corr = (
    corr.where(~np.eye(corr.shape[0], dtype=bool))
    .abs()
    .stack()
    .sort_values(ascending=False)
    .head(5)
)

failure_rate = azure_eda["failure_event"].mean() * 100
main_missing = missing_summary.sort_values("missing_pct", ascending=False).iloc[0]
main_outlier = outlier_summary.iloc[0]

print("SYNTHESE SEMAINE 1 - BRICE")
print("-" * 60)
print(f"1. La table telemetry contient {telemetry.shape[0]:,} lignes et {telemetry.shape[1]} colonnes.")
print(f"2. La variable la plus touchee par les valeurs manquantes est {main_missing['column']} avec {main_missing['missing_pct']} %.")
print(f"3. La variable avec le plus de valeurs aberrantes selon IQR est {main_outlier['variable']} ({main_outlier['outlier_pct']} %).")
print(f"4. Le taux de panne instantanee observe est de {failure_rate:.4f} %, ce qui suggere un fort desequilibre de classes.")
print("5. Paires de variables les plus correlees :")
for (var_a, var_b), value in top_corr.items():
    print(f"   - {var_a} / {var_b} : {value:.3f}")
```
